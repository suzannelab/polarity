```python
def open_sheet(file_name, data_names=['vert', 'edge', 'face', 'cell']):
    """Open hdf5 file

    Open HDF5 file correspond to t time from dirname directory.

    Parameters
    ----------
    directory : str
        complete directory path
    t : int
        time step
    """
    dsets = hdf5.load_datasets(file_name,
                               data_names=data_names)

    
    sheet = Sheet('ellipse', dsets)
    return sheet
```

```python
with_polarity = pd.read_csv('with_polarity.csv')
without_polarity = pd.read_csv('without_polarity.csv')
```

```python
fig = plt.figure(1,figsize=(20, 10))
#create automatically box plot from data_frame

ar1 = pd.DataFrame(with_polarity['orient'])
ar2 = pd.DataFrame(without_polarity['orient'])

data=pd.concat([ar1, ar2], axis=1, sort=False)


data.boxplot()

for i,d in enumerate(data):
    y = data[d]
    x = np.random.normal(i+1, 0.04, len(y))
    plt.plot(x, y, mfc = ["grey","lightblue"][i], mec='k', ms=7, marker="o", linestyle="None")

plt.savefig('orient.eps', dpi = 300, transparent=True)
plt.savefig("orient.pdf", dpi = 300, transparent=True)
```
